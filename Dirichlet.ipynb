# Dirichlet test

```python
import sys
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
q = np.random.dirichlet((1.0, 1.0), size=100000000).transpose()
t = q[0]+q[1]
print(t)
```

```python
x0 = q[0]

h0 = np.histogram(x0, bins = 100, range = [0.0, 1.0])
b0 = h0[1] # bins
a0 = 0.5*(b0[0:-1] + b0[1:])
o0 = h0[0] / (len(x0) * (b0[1] - b0[0]))
print(np.mean(x0))

x1 = 1.0 - q[0]

h1 = np.histogram(x1, bins = 100, range = [0.0, 1.0])
b1 = h1[1] # bins
a1 = 0.5*(b1[0:-1] + b1[1:])
o1 = h1[0] / (len(x1) * (b1[1] - b1[0]))
print(np.mean(x1))
print(np.mean(x0) + np.mean(x1))

plt.plot(a0, o0, 'b-')
plt.plot(a1, o1, 'g-')
plt.plot(a0, 0.5*(o0 + o1), 'r^')
```

```python
def ssdir(size):
    q = np.random.exponential(1.0, size=size)
    return q / (q + np.random.exponential(1.0, size=size))

def sdir(size):
    q = -np.log(1.0 - np.random.uniform(size = size))
    _ = -np.log(1.0 - np.random.uniform(size = size))
    q = q/(q + _)
    del _
    return q

x0 = sdir(size=100000000*2)

h0 = np.histogram(x0, bins = 100, range = [0.0, 1.0])
b0 = h0[1] # bins
a0 = 0.5*(b0[0:-1] + b0[1:])
o0 = h0[0] / (len(x0) * (b0[1] - b0[0]))
print(np.mean(x0))

x1 = 1.0 - x0

h1 = np.histogram(x1, bins = 100, range = [0.0, 1.0])
b1 = h1[1] # bins
a1 = 0.5*(b1[0:-1] + b1[1:])
o1 = h1[0] / (len(x1) * (b1[1] - b1[0]))
print(np.mean(x1))
print(np.mean(x0) + np.mean(x1))

plt.plot(a0, o0, 'b-')
plt.plot(a1, o1, 'g-')
plt.plot(a0, 0.5*(o0 + o1), 'r^')
```
